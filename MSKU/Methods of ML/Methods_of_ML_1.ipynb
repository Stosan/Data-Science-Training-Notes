```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
from scipy import stats
import numpy as np

#creating data
x = np.array([90, 120, 150, 130, 170])
y = np.array([1000, 1300, 1800, 1200, 1380])

#creating OLS regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

def linefitline(b):
    return intercept + slope * b

line1 = linefitline(x)

#plot line # RSS
plt.scatter(x,y)
plt.plot(x,line1, c = 'g')
plt.show()
```

```python
# TSS
line2 = np.full(5,[y.mean()])
plt.scatter(x,y)
plt.plot(x,line2, c = 'r')
plt.show()
```

```python
# RSS
differences_line1 = linefitline(x) - y
line1sum = 0
for i in differences_line1:
    line1sum = line1sum + (i*i)
line1sum
```

```python
# TSS
differences_line2 = line2 - y
line2sum = 0
for i in differences_line2:
    line2sum = line2sum + (i*i)
line2sum
```

```python
# 1 - RSS/TSS
# belirlilik katsayısı
r2 = r2_score(y, linefitline(x))
print('The rsquared value is: ' + str(r2))
```

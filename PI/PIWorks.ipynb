```python
import pandas as pd
import numpy as np
import seaborn as sns

def netflix(p, n, netflixprior=0.5):
    """
P(Y) = probability of n people saying yes

P(F) = probability of the movie being listed on Netflix (prior)

We are looking for the probability of movie being on the Netflix given n people saying yes. P(F | Y) 

P(F | Y) = P(Y | F) P(F) / P(Y)

P(Y | F)  = (1 - p) ** n        "If there is a movie, the probability of saying that n people also"
P(Y | nF)= p**n                 "If there is a movie, the probability of saying that n people not"
P(Y) = (P(Y | F)  * P(F)) + (P(Y | nF) * (1 - P(F)))

    """
    PF= netflixprior
    PYlF = (1 - p) ** n  
    PYlnF = p**n 
    PY= (PYlF * PF) + (PYlnF * (1 - PF))

    p = (PF * PYlF) / PY
    
    return p

P = np.linspace(0,1,100)
N = list(range(1,100))
nn, pp = np.meshgrid(N, P)
zz = np.array([netflix(p, n) for n,p in zip(nn.ravel(), pp.ravel())]).reshape(nn.shape)

df = pd.DataFrame(zz, columns=N, index=P)


sns.heatmap(df)
```

```python
sns.heatmap(df > 0.95)
```

* P > 0.95 is the feasible area

## Pairgrid

```python
g = sns.PairGrid(tips)
```

```python
g = sns.PairGrid(tips)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot)
```

```python
g = sns.PairGrid(tips)
g.map_lower(sns.scatterplot)
```

```python
g = sns.PairGrid(tips)
g.map_diag(sns.histplot)
```

```python
g = sns.PairGrid(tips)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot, color = 'red')
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, color = 'orange')
```

```python
g = sns.PairGrid(tips, corner=True)
g.map_upper(sns.scatterplot, color = 'red')
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, color = 'orange')
```
